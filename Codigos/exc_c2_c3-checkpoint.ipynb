## Centroide de un tri√°ngulo dado los 3 vectores de los vertices.

```python
vertices = np.array([[ 2,  1], #punto 1
                     [-2,  2], # punto 2
                     [-8, 1]]) #punto 3
a,b,c = vertices


for i in range(3):
    if i+1 == 3:
        plt.plot(*vertices[[0, i]].T)
        plt.scatter(*vertices[i], label = "punto {}".format(i+1))
        plt.legend()
        break
    plt.scatter(*vertices[i], label = "punto {}".format(i+1)) # Vertices
    plt.plot(*vertices[[i, i+1]].T)
    

# Centroide
cen = (a+b+c)/3
plt.scatter(*cen)
plt.annotate("Centroide \n ({},{})".format(*cen.round(1)), cen + (0.1,0.1))
```

# Ejercicio 13

Campo vectorial: $F = \left[\begin{matrix}- \frac{y}{x^{2} + y^{2}}\\\frac{x}{x^{2} + y^{2}}\\0\end{matrix}\right]$

Camino parametrizado: $r = \left[\begin{matrix}\cos{\left(t \right)}\\\sin{\left(t \right)}\\- \sin{\left(t \right)} + \cos{\left(t \right)}\end{matrix}\right]
$

### Integral:
$
\int_{0}^{\pi} \left[\begin{matrix}- \frac{y}{x^{2} + y^{2}}\\\frac{x}{x^{2} + y^{2}}\\0\end{matrix}\right] \cdot \left[\begin{matrix}- \sin{\left(t \right)}\\\cos{\left(t \right)}\\- \sin{\left(t \right)} - \cos{\left(t \right)}\end{matrix}\right] \,dt
$

```python
#Plano z = x - y 
#t = np.linspace(0, 2*np.pi, 50)
t = sp.Symbol("t")
x,y,z = sp.symbols("x y z")
#r = np.array([np.cos(t), np.cos(t), np.sin(t)])
r = sp.Matrix([sp.cos(t), sp.sin(t), sp.cos(t)-sp.sin(t)])
dr = sp.diff(r, t)

f1, f2 = -y/(x**2 + y**2), x/(x**2 + y**2)
F = sp.Matrix([f1, f2, 0])

r0 = r.subs(t, 0)
r1 = r.subs(r, sp.pi)

#Integral de F(r)¬∞r'(t)dt
integral = sp.integrate(F.subs({x:r[0], y:r[1]}).dot(dr), (t, 0, sp.pi)) 
print("Integral Horaria (0, pi): {}".format(integral))

r1 = r.subs(r, -sp.pi)

integral = sp.integrate(F.subs({x:r[0], y:r[1]}).dot(dr), (t, 0, sp.pi)) 
print("Integral Anti-Horaria (0, -pi): {}".format(integral))
```

#### Las dos integrales dan los mismo, lo que tiene sentido debido a que el campo vectorial $\vec{F}$ es conservativo, ya que $\nabla \times F = 0$
